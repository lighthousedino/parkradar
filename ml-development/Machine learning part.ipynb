```python
import numpy as np
import pandas as pd
import keras
import matplotlib.pyplot as plt
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import adam_v2
import random
```

```python
FILENAME = 'Updated_data.csv'
DELIMITER = ','

data = np.genfromtxt(FILENAME, delimiter = DELIMITER)



for i in range(0,8):
    new_x = []
    for j in range(len(data)):
        new_x.append([data[j, 0], data[j, 1]])
    new_x = np.array(new_x)

    x = new_x
    y = data[:, 2]

    idx = np.random.permutation(range(len(y)))
    x, y = x[idx,:], y[idx]

    model = Sequential()
    model.add(Dense(256, input_dim=2, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='linear'))
    model.compile(loss='mse', optimizer= adam_v2.Adam(learning_rate=0.001))

    history = model.fit(x, y, epochs = 100, batch_size = 20, verbose = 2)

    model.save(f"carpark {i}")
```

```python
hello =[]
for i in np.arange(0, 24, 0.1):
    input = np.array([1, i]).reshape(1, 2)
    hello.append(np.ravel(model(input)))


plt.plot(hello)
```
